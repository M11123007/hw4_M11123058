```python
invoice=df['INVOICE_NO'].value_counts()
invoice_df=invoice.to_frame()
invoice_df=invoice_df[invoice_df['INVOICE_NO'] > 1]

invoice_df=invoice_df.index
invoice_df=invoice_df.astype(str)

trade_lst_count=invoice_df.tolist()
trade_lst_count
```

```python
trade_lst=[]
df_cache = pd.DataFrame()
before_t = datetime.datetime.now()
for trade in range(len(trade_lst_count)):
#for trade in range(5):
#     print(trade)
    df_cache = df_cache.iloc[0:0]
#     print(invoice_array[trade])
    df_cache=df[df['INVOICE_NO'].astype(str)==trade_lst_count[trade]]
    df_cache=df_cache['ITEM_ID'].astype(str)
    
    array_cache=['']
    array_cache=numpy.unique(df_cache.to_numpy())
    
    trade_lst.append(array_cache.tolist())

after_t=datetime.datetime.now()
runtime_t=after_t-before_t
print(runtime_t)
```

```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import datetime
with open('test.pickle', 'rb') as f:
    trade_lst = pickle.load(f)

print(trade_lst)
```

```python
# from apyori import apriori
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
```

```python
#　https://www.kaggle.com/code/sangwookchn/association-rule-learning-with-scikit-learn
#　https://artsdatascience.wordpress.com/2019/12/10/python-%E5%AF%A6%E6%88%B0%E7%AF%87%EF%BC%9Aapriori-algorithm/
sp1=0.002
sp2=0.003
sp3=0.005


ap=TransactionEncoder()
ap_array=ap.fit(trade_lst).transform(trade_lst)
ap_df=pd.DataFrame(ap_array,columns=ap.columns_)

before_ap1 = datetime.datetime.now()
rules_ap1 = apriori(ap_df,min_support = sp1,use_colnames=True)
after_ap1 = datetime.datetime.now()
runtime_ap1=after_ap1-before_ap1
length_lst=rules_ap1['itemsets']

length_trade_lst=[]
for num in range(len(length_lst)):
    length_trade_lst.append(len(length_lst[num]))
length_trade_lst
rules_ap1['length']=length_trade_lst
```

```python
rules_ap1 = rules_ap1[rules_ap1['length'] > 1]
rules_ap1
```

```python
# fp-growth
# https://www.kaggle.com/code/rjmanoj/fp-growth-algorithm-frequent-itemset-pattern/notebook
from mlxtend.frequent_patterns import fpgrowth
```

```python
te = TransactionEncoder()
te_ary = te.fit(trade_lst).transform(trade_lst)
fp_df = pd.DataFrame(te_ary, columns=te.columns_)

before_fp1=datetime.datetime.now()
rules_fp1=fpgrowth(fp_df,min_support = sp1,use_colnames=True)
after_fp1=datetime.datetime.now()
runtime_fp1=after_fp1-before_fp1


length_fp1_lst=[]
length_fp1=rules_fp1['itemsets']
for num in range(len(length_fp1)):
    length_fp1_lst.append(len(length_fp1[num]))


rules_fp1['length']=length_fp1_lst

rules_fp1=rules_fp1[rules_fp1['length'] > 1]
rules_fp1
```

```python
#Second Support:0.003
# Apriori


before_ap2 = datetime.datetime.now()
rules_ap2 = apriori(ap_df,min_support = sp2,use_colnames=True)
after_ap2 = datetime.datetime.now()
runtime_ap2=after_ap2-before_ap2
length_ap2=rules_ap2['itemsets']

length_ap2_lst=[]
for num in range(len(length_ap2)):
    length_ap2_lst.append(len(length_ap2[num]))

    
rules_ap2['length']=length_ap2_lst

rules_ap2 = rules_ap2[rules_ap2['length'] > 1]
rules_ap2
```

```python
# FP-growth

before_fp2=datetime.datetime.now()
rules_fp2=fpgrowth(fp_df,min_support = sp2,use_colnames=True)
after_fp2=datetime.datetime.now()
runtime_fp2=after_fp2-before_fp2


length_fp2_lst=[]
length_fp2=rules_fp2['itemsets']
for num in range(len(length_fp2)):
    length_fp2_lst.append(len(length_fp2[num]))


rules_fp2['length']=length_fp2_lst

rules_fp2=rules_fp2[rules_fp2['length'] > 1]
rules_fp2
```

```python
#Third Support:0.005
# Apriori


before_ap3 = datetime.datetime.now()
rules_ap3 = apriori(ap_df,min_support = sp3,use_colnames=True)
after_ap3 = datetime.datetime.now()
runtime_ap3=after_ap3-before_ap3
length_ap3=rules_ap3['itemsets']

length_ap3_lst=[]
for num in range(len(length_ap3)):
    length_ap3_lst.append(len(length_ap3[num]))

    
rules_ap3['length']=length_ap3_lst

rules_ap3 = rules_ap3[rules_ap3['length'] > 1]
rules_ap3
```

```python
# FP-growth

before_fp3=datetime.datetime.now()
rules_fp3=fpgrowth(fp_df,min_support = sp3,use_colnames=True)
after_fp3=datetime.datetime.now()
runtime_fp3=after_fp3-before_fp3


length_fp3_lst=[]
length_fp3=rules_fp3['itemsets']
for num in range(len(length_fp3)):
    length_fp3_lst.append(len(length_fp3[num]))


rules_fp3['length']=length_fp3_lst

rules_fp3=rules_fp3[rules_fp3['length'] > 1]
rules_fp3=rules_fp3.drop(['length'], axis=1)
rules_fp3
```

```python
print("Run Time Apriori1",runtime_ap1,"Support:",sp1)
print("Run Time FP-Growth1",runtime_fp1,"Support:",sp1)
print("Run Time Apriori2",runtime_ap2,"Support:",sp2)
print("Run Time FP-Growth2",runtime_fp2,"Support:",sp2)
print("Run Time Apriori3",runtime_ap3,"Support:",sp3)
print("Run Time FP-Growth3",runtime_fp3,"Support:",sp3) 
```

```python
from mlxtend.frequent_patterns import association_rules

rules_ap1 = apriori(ap_df,min_support = sp1,use_colnames=True)
AP1_association_rules=association_rules(rules_ap1,metric="lift")
rules_ap2 = apriori(ap_df,min_support = sp2,use_colnames=True)
AP2_association_rules=association_rules(rules_ap2)
rules_ap3 = apriori(ap_df,min_support = sp3,use_colnames=True)
AP3_association_rules=association_rules(rules_ap3)
rules_fp1=fpgrowth(fp_df,min_support = sp1,use_colnames=True)
FP1_association_rules=association_rules(rules_fp1)
rules_fp2=fpgrowth(fp_df,min_support = sp2,use_colnames=True)
FP2_association_rules=association_rules(rules_fp2)
rules_fp3=fpgrowth(fp_df,min_support = sp3,use_colnames=True)
FP3_association_rules=association_rules(rules_fp3,support_only=False)
```

```python
AP1_association_rules
```

```python
# AP1
AP1_item=input("AP1 Rule:")


# https://linuxhint.com/use-frozenset-objects-in-python/
AP1Results=AP1_association_rules[AP1_association_rules['antecedents']=={AP1_item}]

AP1Results_lst=AP1Results['consequents'].tolist()

AP1Results_df=pd.DataFrame()
for set in range(len(AP1Results_lst)):
    AP1Results_df=AP1Results_df.append(list(AP1Results_lst[set]))
    
AP1Results_ar=numpy.unique(AP1Results_df.to_numpy())

print("可能購買的品項ID有:")
for item in range(len(AP1Results_ar)):
    print(AP1Results_ar[item])
```
